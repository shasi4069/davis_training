```python
#NumericCalculation.py
#function to add two number
def calculateSum(x,y):
  return(x+y)
#function to sub two number
def calculateSub(x,y):
  return(x-y)
#function to mul two number
def calculateMul(x,y):
  return(x*y)
#function to div two number
def calculateDiv(x,y):
  return(x/y)
```
