```python
def atm():
    balance = 10000

    while True:
        amount = float(input("Enter amount (0 to exit): "))

        if amount == 0:
            break

        if amount < 0:
            print("Invalid amount")
        elif amount > balance:
            print("Insufficient balance")
        else:
            balance -= amount
            print("Withdrawal successful")
            print("Remaining balance:", balance)

atm()
```
