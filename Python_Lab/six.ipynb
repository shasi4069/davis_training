```python
def calculate_bonus(salary, years):
    if years > 10:
        bonus = salary * 0.20
    elif years >= 5:
        bonus = salary * 0.10
    else:
        bonus = salary * 0.05

    return bonus

for i in range(3):
    salary = float(input("Enter salary: "))
    years = int(input("Enter experience: "))

    bonus = calculate_bonus(salary, years)
    print("Bonus:", bonus)
```
