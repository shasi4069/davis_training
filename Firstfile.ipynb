```python
#to find  out oput hcf
#input two number
num1=int(input("Enter first number:"))
num2=int(input("Enter second number:"))
#-----------------------------------------
if(num2==0):
  print("HCF :",num1)
else:
    while(num2!=0):
      rem=num1%num2
      num1=num2
      num2=rem
      #-------------------------
      print("HCF :",num1)
```

```python
# find out greatest digit present in given number
# input : 3723
#output: greatest digit - 7
num = input("Enter a number: ")

greatest = 0

for digit in num:
    if int(digit) > greatest:
        greatest = int(digit)

print("Greatest digit -", greatest)
```

```python
# to find out factoriazl of a number
num=int(input("enter a number :"))
if(num<0):
  print("factorial not possible")
elif(num==0):
  print("factorial=1")
else:
  fact=1
  for x in range(1,num+1):
    fact=fact*x
  print("factorial=",fact)
```
